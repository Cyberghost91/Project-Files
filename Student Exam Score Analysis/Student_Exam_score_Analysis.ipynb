```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Student_Exam_Score.csv")
print(df.head(10))
```

```python
df = df.drop("Unnamed: 0", axis=1)
df.head()
```

# Gender Distribution

```python
plt.figure(figsize=(5,6))
ax = sns.countplot(data=df, x= "Gender")
ax.bar_label(ax.containers[0]) 
plt.show()
```

```python
# It clearly shows that the no. of females are more than the no. of male from above mentioned chart analysis.
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,
            annot=True
           )
plt.title("Parents Education and students score")
plt.show()
```

```python
# From the above chart we have concluded that there is impact on the student scores due to there
# parents Education
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1, 
            annot=True
           )
plt.title("Parents marital status and students score")
plt.show()
```

```python
# From the above chart we have concluded that there is no/negligible impact on the student scores due to there
# parents marital status
```

```python
plt.figure(figsize=(3,3))
sns.boxplot(data = df, x="MathScore")
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.boxplot(data = df, x="ReadingScore")
plt.show()
```

# Distribution of ethnic group

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

list = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]

L = ["groupA", "groupB", "groupC", "groupD", "groupE"]
plt.pie(list,
       labels=L,
        autopct= "%1.2f%%"
       )
plt.title("Distribution of Ethnic Groups")
plt.show()
#print(groupA)
```
